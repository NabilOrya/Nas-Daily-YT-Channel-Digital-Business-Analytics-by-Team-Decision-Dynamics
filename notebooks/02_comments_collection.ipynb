```python
import os
import sys
import pandas as pd

project_root = os.path.abspath("..")
if project_root not in sys.path:
    sys.path.append(project_root)

from src.api_client import get_youtube_client
```

```python
df_videos = pd.read_csv("../data/raw/videos/only_1_minute_videos.csv")
video_ids = df_videos["video_id"].tolist()

youtube = get_youtube_client()
```

```python
# Comment fetching logic is implemented in `src/comments.py`.
# We keep this cell as documentation only and use the shared function
# via `from src.comments import fetch_comments` in the next cell.

```

```python
from src.comments import fetch_comments
from tqdm.auto import tqdm

all_comments = []

for vid in tqdm(video_ids):
    try:
        all_comments.extend(fetch_comments(youtube, vid))
    except Exception as e:
        print(f"Error on video {vid}: {e}")

df_comments = pd.DataFrame(all_comments)
df_comments.head()
```

```python
os.makedirs("../data/raw/comments", exist_ok=True)

df_comments.to_csv(
    "../data/raw/comments/only_1_minute_comments.csv",
    index=False
)
```
